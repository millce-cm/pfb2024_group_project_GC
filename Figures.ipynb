# Create a DataFrame from our dictionary

```python
df = pd.DataFrame(Whole_Genome)
df
```

```python
##transpose from wide to long. 
Whole_Genome_Long = df.transpose()
Whole_Genome_Long
```

# Order the values based on chromosome #

```python
custom_order = [str(i) for i in range(1, 23)] + ['X', 'Y']
custom_order
```

```python
# Reset the index and create a new column
Whole_Genome_Long.reset_index(inplace=True)

# Now, rename the new first column (which was the index)
Whole_Genome_Long.rename(columns={'index': 'Chromosome'}, inplace=True)

# Check the DataFrame
Whole_Genome_Long
```

```python
Whole_Genome_Long['Chromosome'] = pd.Categorical(Whole_Genome_Long['Chromosome'], categories=custom_order, ordered=True)

# Sort the DataFrame by the categorical column
Whole_Genome_Long_Sorted = Whole_Genome_Long.sort_values('Chromosome')
```

# Plot GC Content for each chromosome, with and without Ns 

```python
#Put the two figures side-by-side:
# Create a figure with 2 subplots (2 columns, 1 row)
# Sort the DataFrame by 'GC_content_all'
Whole_Genome_Long_Sorted_PanelA = Whole_Genome_Long_Sorted.sort_values(by='GC_content_all')

# Sort the DataFrame by 'GC_content_non'
Whole_Genome_Long_Sorted_PanelB = Whole_Genome_Long_Sorted.sort_values(by='GC_content_non')

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

Whole_Genome_Long_Sorted_PanelA.plot(x='Chromosome', y='GC_content_all', kind='bar', ax=axs[0])
axs[0].set_title('GC Content in the Human Genome (including Ns)')
axs[0].set_xlabel('Chromosome')
axs[0].set_ylabel('GC Content')

# Plot the second bar plot
Whole_Genome_Long_Sorted_PanelB.plot(x='Chromosome', y='GC_content_non', kind='bar', ax=axs[1])
axs[1].set_title('GC Content in the Human Genome (not including Ns)')
axs[1].set_xlabel('Chromosome')
axs[1].set_ylabel('GC Content')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

# Calculate the average GC content across all chromosomes
# Plot those values side-by-side for with and without Ns

```python
#Create the average values
avg_gc_content_all = Whole_Genome_Long_Sorted['GC_content_all'].mean()
avg_gc_content_non = Whole_Genome_Long_Sorted['GC_content_non'].mean()
```

```python
# Create a figure
fig, ax = plt.subplots(figsize=(5, 5))

# Bar width
bar_width = 0.5

# X locations for the bars
x = ['GC_content_all', 'GC_content_non']

# Average bar heights
average_values = [avg_gc_content_all, avg_gc_content_non]

# Create bars for average values
bars = ax.bar(x, average_values, width=bar_width, color=['lightblue', 'lightblue'], label='Average GC Content')

# Overlay individual data points
# Use a scatter plot for individual points for each category:
for i, col in enumerate(['GC_content_all', 'GC_content_non']):
    ax.scatter([x[i]] * len(Whole_Genome_Long_Sorted), Whole_Genome_Long_Sorted[col], color='blue', alpha=0.5, label='Individual Data Points' if i == 0 else "")

# Set titles and labels:
ax.set_title('Average GC Content Across the Human Genome')
ax.set_ylabel('GC Content')
#ax.legend()

# Show the plot:
plt.tight_layout()
plt.show()
```

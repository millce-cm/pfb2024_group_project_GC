```python
#!/usr/bin/env python3
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('/Users/pfb2024/PFB_problemsets/Project/pfb2024_group_project_GC/fasta_dict_cds.p', 'rb') as file:
    x = pickle.load(file)
x
```

```python
data_frame_x_cds= pd.DataFrame.from_dict(x)
data_frame_x_cds
```

```python
data_frame_x_cds_sorted_GCwn = data_frame_x_cds.sort_values('gc_percentage')
# data_frame_x.chromosome = pd.to_numeric(data_frame_x.chromosome, errors='coerce')
# data_frame_x.sort_values('chromosome')
data_frame_x_cds_sorted_GCwn
```

```python
ax = data_frame_x_cds_sorted_GCwn.plot(kind='bar', title ="GC content of each human chromosome with N nucleotides",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("Chromosome",fontsize=12)
ax.set_ylabel("GC % (with n)",fontsize=12)
```

```python
ax = data_frame_x_cds_sorted_GCwn.gc_percentage.plot(kind='bar', title ="GC content of each human chromosome with N nucleotides",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("Chromosome",fontsize=12)
ax.set_ylabel("GC % (with n)",fontsize=12)
plt.ylim((40,60))
```
